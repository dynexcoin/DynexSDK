# CFQRBM Image Reconstruction using an QRBM based Collaborative Filtering

Fast and accurate approach to train Restricted Boltzmann Machine for Collaborative Filtering using Quantum Annealing, running on the Dynex Neuromorphic Computing Platform.

Based on the bars and stripes dataset

```python
import numpy as np
import timeit
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook as tqdm

from CFQIRBM.qrbm import QRBM

image_height = 6 
image_width = 6 

len_x = image_height * image_width
len_y = 0

f = open("datasets/6x6.txt")
data = [[int(data) for data in f.read() if data in "01"]]
```

```python
img = np.array(data).reshape(image_height,-1)
plt.imshow(img)
```

## Construt the CFQIRBM

```python
bm = QRBM(n_visible=len_x, n_hidden=56)
bm.image_height = image_height
bm.tqdm = tqdm
```

## Train the CFQIRBM

```python
bm.train(data, len_x, len_y, epochs = 50, num_samps = 1, lr = 1000, decay = 0.01, epochs_to_test = 1, print_training_data=True)
```
